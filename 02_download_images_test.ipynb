```python
import json
from IPython.display import Image
```

```python
with open('data/metadata/edan_anderson_photos.json','r') as json_in:
    anderson_json = json.load(json_in)
```

```python
anderson_sample = anderson_json[0]
print(json.dumps(anderson_sample, indent=2))
```

```python
if 'online_media' in anderson_sample['content']['descriptiveNonRepeating']:
    edan_id = anderson_sample['content']['descriptiveNonRepeating']['record_ID']
    ids_id = anderson_sample['content']['descriptiveNonRepeating']['online_media']['media'][0]['idsId']
    print(edan_id, ids_id)
```

```python
def si_iiif_url(ids_id, region='full', size='500,', rotation='0', 
                     quality='default', format='jpg'):
    SI_ENDPOINT = 'https://ids.si.edu/ids/iiif'
    si_url = f"{SI_ENDPOINT}/{ids_id}/{region}/{size}/{rotation}/{quality}.{format}"
    return si_url
```

```python
print(si_iiif_url(ids_id))
```
